First let´s explore a bit this dataset in order to understand its structure

```python
import pandas as pd
%matplotlib inline
recent_grads=pd.read_csv("recent-grads.csv")
recent_grads.iloc[0]
```

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

* Rank - Rank by median earnings (the dataset is ordered by this column).
* Major_code - Major code.
* Major - Major description.
* Major_category - Category of major.
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time.
* Men - Male graduates.
* Women - Female graduates.
* ShareWomen - Women as share of total.
* Employed - Number employed.
* Median - Median salary of full-time, year-round workers.
* Low_wage_jobs - Number in low-wage service jobs.
* Full_time - Number employed 35 hours or more.
* Part_time - Number employed less than 35 hours.

## 2. SCATTER PLOTS

Now we will display some scatter plots to identify some relations between the columns and obtain some interesting findings. However, let´s to limit x axis number since thre are smaple size extremely large and most of them are small sample size and poblation in general.

```python
ax1=recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
ax1.set_xlim(0, 500)
ax1.set_ylim(18000, 120000)
```

How we can see there is a low negtive correlation, it means the smallest sample size trend to have a better median salary of full-time

```python
ax2=recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
ax2.set_ylim(0, 0.20)
ax2.set_xlim(0, 1000)
```

as the previous scatter correlation is negative 

```python
ax3=recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time')
ax3.set_ylim(2000, 120000)
ax3.set_xlim(2000, 50000)
```

while median salary increases number of people employed as full time dicrease

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. ShareWomen') 
```

there is no correlation betweenunemployment rate and women as share of total

```python
ax4=recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs. Men')
ax4.set_ylim(2000, 120000)
ax4.set_xlim(2000, 50000)
```

correlation negative, while median salary goes up number of male graduates goes down

```python
ax5=recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs. Men')
ax5.set_ylim(2000, 120000)
ax5.set_xlim(2000, 50000)
```

the same case as previous scatter

## 3. HISTOGRAMS

Now will explore the distribution of values in the different columns

```python
ax1=recent_grads['Sample_size'].hist(bins=25,range=(0,500))
ax1.set_title("Sample_size")
```

the histogram shows the most of sample sizes are the smallest

```python
ax1=recent_grads['Median'].hist(bins=25,range=(0,60000))
ax1.set_title("Median")
```

the highest median salary is concreted between 30.000 t0 40.000

```python
ax1=recent_grads['Employed'].hist(bins=25,range=(0,8000))
ax1.set_title("Employed")
```

the highest number employed is concentrated between 500 to 1000

```python
ax1=recent_grads['Full_time'].hist(bins=25,range=(0,15000))
ax1.set_title("Full_time")
```

It is most frequent that 2000 people have full time job in pur dataset, while number of employed increases less are the people have a full job

```python
ax1=recent_grads['ShareWomen'].hist(bins=25)
ax1.set_title("ShareWomen")
```

as we can see women as share of total is almost 80% being this the data most frequently

```python
ax1=recent_grads['Unemployment_rate'].hist(bins=25)
ax1.set_title("Unemployment_rate")
```

6% of unemployment is most frequently in this dataset for recent graduates

## 5. COMPARISON GROUPED BAR PLOTS

Let´s display bar plot of the first 10 values for sharewomen column and the last 10 values. Also will display first 10 values for unemployment rate column and the last 10 values

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
recent_grads[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate', legend=False)
recent_grads[163:].plot.bar(x='Major', y='Unemployment_rate', legend=False)
```
